```python
# import numpy as np
# import pandas as pd

# import bokeh
# import bokeh.io
# import bokeh.plotting

# bokeh.io.output_notebook()
import numpy as np
import pandas as pd
import matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filename = 'nfl_matrix.csv'
data = pd.read_csv(filename, index_col=None, header=None)
data = data.values

for i in range(0, len(data)):
    for j in range(0, len(data)):
        if i > j or i == 1 or j == 1:
            data[i][j] = -1
data
```

```python
# create a mask for values < 0
mask = np.ones_like(data)
mask[np.triu_indices_from(mask)] = False
# mask where row == 1
mask[1, :] = True
mask[:, 1] = True
# mask where value == 0
mask[data == 0] = True

# make it bigger
plt.figure(figsize=(15, 15))

p = sns.heatmap(data, square=True, cmap='magma_r', cbar=False, mask=mask, annot=False)
# put x axis at the top
p.xaxis.tick_top()
# rotate the x axis labels
plt.xticks(rotation=90)

plt.show()
```
